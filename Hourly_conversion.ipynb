```python
import xarray as xr

path = '/DISK-0/gokul/ENGY/'
storm = 'dat2/burevi'

# Load ERA5 hourly data
ds = xr.open_dataset(path+storm+'.nc')

# Select every 3th hour
ds_3hourly = ds.sel(time=ds['time.hour'] % 3 == 0)

# If you want to cut any time steps you can use the code
#ds_3hourly_modified = ds_3hourly.isel(time=slice(1, None))

# Save the 3-hourly data to a new NetCDF file
ds_3hourly.to_netcdf(path+storm+'_3.nc')

dataset = xr.open_dataset(path+storm+'_3.nc')
time = dataset.variables['time']
print(f"time steps {time}")
print(f"No.of timesteps {len(time)}")

```
